## Mutual Information

```python
def MIPick(tr,te):
    MIList = []
    indexList =[]

    for ind in (te.index):
        training = tr
        testing = te
        trainingAdded = pd.concat([training, testing.iloc[[ind]]], axis=0)
        indexUse = testing.iloc[[ind]].index
        testing = testing.drop(indexUse)

        trainingX = training[training.columns[:-1]].to_numpy()
        trainingY = training[training.columns[-1]].to_numpy()
        
        trainingXAdd = trainingAdded[trainingAdded.columns[:-1]].to_numpy()
        trainingYAdd = trainingAdded[trainingAdded.columns[-1]].to_numpy()
        
        rf = RandomForestClassifier()
        testingX = testing[testing.columns[:-1]].to_numpy()
        testingY = testing[testing.columns[-1]].to_numpy()
        # when the label is 0
        trainingYAdd[-1] = 0

        rf.fit(trainingXAdd, trainingYAdd)
        pred0List = rf.predict_proba(testingX)
        score0 = sum([max(i) for i in pred0List])
        
        # when label is 1

        trainingYAdd[-1] = 1

        rf.fit(trainingXAdd, trainingYAdd)
        pred1List = rf.predict_proba(testingX)
        score1 = sum([max(i) for i in pred1List])

        maximum = max(score0, score1)
        
        # training without concatenating
        
        
        rf.fit(trainingX, trainingY)
        predList = rf.predict_proba(testingX)
        score = sum([max(i) for i in predList])
        
        # mutual information value 
        
        value = score - maximum
        
        MIList.append(value)
        indexList.append(ind)

    indexPick = indexList[MIList.index(min(MIList))]
    return indexPick




```

```python

trainX = dataTrainFeature
trainY = dataTrainLabel
testX = dataTestFeature
testY = dataTestLabel

Train = pd.concat([trainX, trainY],axis=1)
Test = pd.concat([testX, testY], axis = 1)
```

```python
start = time.time()
accuracyTrainAverageMI = []
accuracyTestAverageMI = []
accuracyTrainSDMI = []
accuracyTestSDMI = []

train = Train
test = Test

chosenMI = []

for random in range(30):


    # implement MI

    pickMIRow = MIPick(train, test)
    print(pickMIRow)
    
    train = pd.concat([train, test.iloc[[pickMIRow]]], axis=0)
    test.drop(pickMIRow)
    chosenMI.append(pickMIRow)


    lossTrainSubset = []
    lossTestSubset = []

    for i in range(5):  # 5 times for different random seeds
        # to have 5 folds to do cross validation

        kFold = KFold(n_splits = 5, shuffle = True, random_state = i+10)


        # run across 5 iterations / 5 folds
        predictions = []
        actual = []


        trainX = train[train.columns[:-1]].to_numpy()
        trainY = train[train.columns[-1]].to_numpy()

        testX = test[test.columns[:-1]].to_numpy()
        testY = test[test.columns[-1]].to_numpy()

        for trainIndex, valIndex in kFold.split(trainX):
            xTrain, xVal = trainX[trainIndex], trainX[valIndex]
            yTrain, yVal = trainY[trainIndex], trainY[valIndex]
            rf = RandomForestClassifier()
            rf.fit(xTrain, yTrain)
            yPred = rf.predict(xVal)
            predictions += yPred.tolist()
            actual += yVal.tolist()

        loss = accuracy_score(actual,predictions)
        lossTrainSubset.append(loss)

        rf.fit(trainX, trainY)
        predictTest = rf.predict(testX)
        lossTest = accuracy_score(predictTest, testY.tolist())

        lossTestSubset.append(lossTest)

    accuracyTrainAverageMI.append(np.mean(lossTrainSubset))
    accuracyTrainSDMI.append(np.std(lossTrainSubset))
    print(np.mean(lossTrainSubset))
    print(np.std(lossTrainSubset))
    accuracyTestAverageMI.append(np.mean(lossTestSubset))
    accuracyTestSDMI.append(np.std(lossTestSubset))
end = time.time()
elapsed = end-start
print(elapsed)
```
